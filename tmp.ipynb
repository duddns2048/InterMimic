```python
import torch
num=1
data_list = ["./parahome/s6_8_cup,diningtable,desk_110.pt",
"./parahome/s6_10_book,desk,bookshelf_111.pt",
"./parahome/s7_5_pan,gasstove,sink_110.pt",
"./parahome/s7_5_pan,sink,gasstove_111.pt",
"./parahome/s10_7_chair,desk,diningtable_110.pt",
"./parahome/s10_14_kettle,sink,diningtable_110.pt"]

data_path = data_list[num]
data = torch.load(data_path)

data.keys()
num_frame = data['data'].shape[0]
intermimic_data = torch.zeros(num_frame, 526+65)
intermimic_data[:,0:3] = data['data'][:,0:3] # root pos

# intermimic_data[:,3:7] = data['data'][:,3:7]
intermimic_data[:,3:7] = data['data'][:,3:7][:,[1,2,3,0]] # root rot

intermimic_data[:,7:9] = torch.zeros(num_frame,2) # unknown
intermimic_data[:,9:162] = data['data'][:,9:162]  # 3 * 51 human joint poses
intermimic_data[:,162:318] = data['data'][:,162:318] # 3 * 52 global human joint positions (including pelvis)
intermimic_data[:,318:321] = data['data'][:,526:529] # object position
intermimic_data[:,321:325] = data['data'][:,529:533] # object rotation (quaternion)
intermimic_data[:,325:330] = torch.zeros(num_frame,5) # unknown
intermimic_data[:,330:331] = data['data'][:,538:539] # object contact (binary)
intermimic_data[:,331:383] = data['data'][:,539:591] # human contact (52 joints binary)
# intermimic_data[:,383:591] = data['data'][:,318:526]
intermimic_data[:,383:591] = data['data'][:,318:526].reshape(-1,52,4)[:,:,[1,2,3,0]].reshape(-1,52*4) # 4 * 52 human joint quaternion integration of the global and local rotation

save_list =[
 "./parahome/intermimic_data/sub6_parahome_drink_cup_0.pt",
 "./parahome/intermimic_data/sub6_parahome_place_book_0.pt",
 "./parahome/intermimic_data/sub7_parahome_move_pan_short.pt",
 "./parahome/intermimic_data/sub7_parahome_move_pan_long.pt",
 "./parahome/intermimic_data/sub10_parahome_getup_chair_0.pt",
 "./parahome/intermimic_data/sub10_parahome_move_kettle_sink.pt"]
save_path = save_list[num]
torch.save(intermimic_data, save_path)
```
